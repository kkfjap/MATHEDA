```python
# imports and packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

# Current-Power

```python
# Load Dataset
df = pd.read_csv(r'dataset\current-power.csv', # r read code as is
                delimiter=',') #delimiter separates data
df
```

```python
#summary of statistics
df.describe()
```

```python
#covariance of current and power
df.cov()
```

```python
#scatter plot
x = df['Current']
y = df['Power']

plt.figure()
plt.scatter(x, y)
```

```python
#scatter plot
x = df['Current']
y = df['Power']

plt.figure()
plt.scatter(x, y)
plt.title('Device Current vs Power Response')
```

```python
#scatter plot

plt.figure()
plt.scatter(df['Current'], df['Power'] )
plt.title('Device Current vs Power Response')
```

```python
#scatter plot
x = df['Current']
y = df['Power']

plt.figure()
plt.scatter(x, y)
plt.title('Device Current vs Power Response')
plt.ylabel('Power(W)')
plt.xlabel('Current(A)')
plt.show
```

```python
#correlation coefficient of current and power
df.corr()
```

```python
# regression line
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
line = slope * x + intercept

# scatter plot
plt.figure(figsize = (7,4))
plt.scatter(x, y,
            color = '#1C1F24',
            alpha=0.6,
            label='Data Points')
#line plot
plt.plot(x, line,
            color = "#FF3341",
            label=f'Regression Line (r={r_value:.4f})')

plt.title('Device Current vs Power Response with Regression Line')
plt.ylabel('Power(W)')
plt.xlabel('Current(A)')    
plt.legend()
plt.show()
```

# Air-Quality

```python
#Load Dataset
air = pd.read_csv(r"dataset\air-quality.csv",
                  delimiter = ',')
air 
```

```python
#regression line
slope, intercept, r_value, p_value, std_err = stats.linregress(air['NOx'], air['Relative_Humidity'])
line = slope * air['NOx'] + intercept

#scatter plot
plt.figure(figsize = (7,4))
plt.scatter(air['NOx'], air['Relative_Humidity'],
            color = '#1C1F24',
            alpha=0.6,
            label='Data Points')

#line plot
plt.plot(air['NOx'], line,
            color = "#FF3341",
            label=f'Regression Line (r={r_value:.4f})')

plt.title('NOx vs Humidity Correlation')
plt.ylabel('Relative Humidity (%)')
plt.xlabel('NOx (ug/cu.m)')
plt.legend()
plt.show()
```

```python
#regression line
slope, intercept, r_value, p_value, std_err = stats.linregress(air['NO2'], air['Relative_Humidity'])
line = slope * air['NO2'] + intercept

#scatter plot
plt.figure(figsize = (7,4))
plt.scatter(air['NO2'], air['Relative_Humidity'],
            color = '#1C1F24',
            alpha=0.6,
            label='Data Points')

#line plot
plt.plot(air['NO2'], line,
            color = "#FF3341",
            label=f'Regression Line (r={r_value:.4f})')

plt.title('NO2 vs Relative Humidity')
plt.ylabel('Relative Humidity (%)')
plt.xlabel('NO2 (ug/cu.m)')
plt.legend()
plt.show()
```

```python
#regression line
slope, intercept, r_value, p_value, std_err = stats.linregress(air['Temperature'], air['Relative_Humidity'])
line = slope * air['Temperature'] + intercept

#scatter plot
plt.figure(figsize = (7,4))
plt.scatter(air['Temperature'], air['Relative_Humidity'],
            color = '#1C1F24',
            alpha=0.6,
            label='Data Points')

#line plot
plt.plot(air['Temperature'], line,
            color = "#FF3341",
            label=f'Regression Line (r={r_value:.4f})')

plt.title('Temperature vs Humidity Correlation')
plt.ylabel('Relative Humidity (%)')
plt.xlabel('Temperature (C)')
plt.legend()
plt.show()
```
