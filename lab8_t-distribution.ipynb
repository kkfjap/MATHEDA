```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#Load Dataset
df = pd.read_csv(r"C:\Users\HP\Desktop\CPE - 2A\2nd Term\old-MATHEDA\dataset\transformer-voltage-10-sample.csv",
                 delimiter=",",
                 index_col = ["Sample"])
df
```

```python
#Summary of Dataframe
df.info()
```

```python
#Summary of Statistics
df.describe()
```

```python
#t-critical for two-tailed tes 
alpha = 0.05
n = df['Voltage'].count()

t_critical = stats.t.ppf(1 - alpha/2, df=n-1)
t_critical
```

```python
#Confidence Interval
N = df['Voltage'].count()
mean = df['Voltage'].mean()
std = df['Voltage'].std(ddof=1)

ci_lower, ci_upper = stats.t.interval(
    confidence = 0.95,
    df = n-1,
   loc = mean,
   scale = std/np.sqrt(n)
)

print(f"CI: {ci_lower:.2f} to {ci_upper:.2f}")
```

```python
#Voltage Distribution
mean = df['Voltage'].mean()
std = df['Voltage'].std(ddof=1)

#x-axis
x_min = df['Voltage'].min() 
x_max = df['Voltage'].max()

#Normal Distribution Plot
x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, mean, std)
plt.plot(x, p, 
         linewidth=2,
            color='blue')

#Histogram
plt.hist(df['Voltage'],
         bins = 5,
         density = True,
         alpha = 0.8,
         color = '#126782')

plt.title('Voltage Distribution')
plt.xlabel('Voltage(V)')
plt.ylabel('Probability Density')

#Confidence Interval
plt.axvline(ci_lower, 
            color='black', 
            linestyle='--')
plt.axvline(ci_upper,
            color='black', 
            linestyle='--')
plt.show()
```

```python
#Mean of SIze 6
mean_size_6 = df.iloc[0,1:13].mean() #exclude column index 13
mean_size_6
```

```python
#Standard Deviation of Size 6
std_size_6 = df.iloc[0,1:13].std()
std_size_6
```

```python
#Size 6 95% CI
n = df.shape[1]-1 #returns 13 in (17,13)

lower_6, upper_6 = stats.t.interval(
    confidence = 0.95,
    df = n-1,
   loc = mean_size_6,
   scale = std_size_6/np.sqrt(n)
)

print(lower_6)
print(upper_6)
```

```python
#95% Confidence Level
mean = df.iloc[:17,1:13].mean(axis = 1)
std = df.iloc[:17,1:13].std(axis = 1)

df['ci_lower'], df['ci_upper'] = stats.t.interval(
    confidence = 0.95,
    df = n-1,
   loc = mean,
    scale = std/np.sqrt(n)
)
df
```

```python
#fill NAN with 0
df['ci_lower'] = df['ci_lower'].fillna(0)
df['ci_upper'] = df['ci_upper'].fillna(0)
df
```
