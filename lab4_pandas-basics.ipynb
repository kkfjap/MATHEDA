# Descriptive Statisitcs

```python
# Jaguar vs Ranger
resistance = np.array([
    [10.6, 9.1, 9.3, 9.8, 10.5, 10.4, 9.5, 11, 10.4, 3, 9.8 ],
    [10.1, 11, 9.1, 20, 9.2, 10.8, 9.9, 9.2, 9.1, 9.1, np.nan]
]).T
resistance
```

```python
#Creating Dataframe from Numpy Array
df = pd.DataFrame(resistance, columns=['Jaguar', 'Ranger'])
df
```

```python
#Summary of Dataframe
df.info()
```

```python
type(df)
```

```python
#SUmmary of Statistics
df.describe()
```

```python
#Median
jag_median = df['Jaguar'].median()
pan_median = df['Ranger'].median()

print(f'Jaguar Median = {jag_median:.2f}ohms')
print(f'Ranger Median = {pan_median:.2f}ohms')
```

```python
#Mode
jag_mode = df['Jaguar'].mode()
pan_mode = df['Ranger'].mode()
```

```python
jag_mode
```

```python
jag_mode.info()
```

```python
type(jag_mode)
```

```python
print('Jaguar Mode ; )')
for i in jag_mode:
    print(i)
```

```python
print('Panther Mode: ')

for i in pan_mode:
    print(i)
```

```python
# Jaguar IQR
Q1 = df['Jaguar'].quantile(0.25)
Q3 = df['Jaguar'].quantile(0.75)

print(f'Jaguar IQR - {Q3-Q1:.2f} ohms')
```

```python
#Panther IQR
Q1 = df['Ranger'].quantile(0.25)
Q3 = df['Ranger'].quantile(0.75)

print(f'Panther IQR = {Q3-Q1:.2f} ohms')
```

# Bar Chart

```python
jag_mean = df['Jaguar'].mean()
pan_mean = df['Ranger'].mean()

plt.figure(figsize=(7,4))
bars = plt.bar(['Jaguar', 'Ranger'], [jag_mean, pan_mean], 
               color='#126782')
plt.title('Average Resistance by Machine')
plt.ylabel('Resistance (ohms)')

#Bar labels
plt.bar_label(bars, padding=3)
plt.margins(y=0.1)
plt.show()
```
